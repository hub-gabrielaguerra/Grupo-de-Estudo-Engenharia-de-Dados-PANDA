Computation of association score: http://www.biomedical-web.com/pirpheno/help.jsp#score

O association score é calculado integrando três fontes de informação: evidências diretas de associações já reportadas entre piRNAs e doenças, similaridade entre doenças (permitindo inferir associações a partir de doenças relacionadas) e similaridade entre piRNAs (permitindo transferir informação entre moléculas semelhantes), resultando em um valor que representa o nível de confiança na associação.

ex. 0.05 - menor suporte/menos evidência
ex. 0.6 - maior suporte/maior evidência
